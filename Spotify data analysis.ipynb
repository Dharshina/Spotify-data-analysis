```python
#Top 10 songs popularity>90
#Slicing till 10
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
#converting ms to seconds
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(10,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#Plotting correlation between loudness(yaxis), energy(xaxis)
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='loudness', x='energy', color='c').set(title="Loudness vs Energy correlation")
```

```python
#Plotting correlation between popularity and Acousticness-Indirect relation(regression line is downwards)
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='popularity', x='acousticness', color='y').set(title="Popularity vs Acousticness correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="No.of songs per year")
```

```python
total_dr = df_tracks.duration
sns.barplot(x=years, y=total_dr)
plt.title('Total Duration by Year')

plt.show()
```

```python
sns.lineplot(x=years, y=total_dr)
plt.title('Total Duration over Years')

plt.show()
```

```python
df_features = pd.read_csv("D:\Dharshina\Jupyter Notebook\Spotify\SpotifyFeatures.csv")
df_features
```

```python
sns.barplot(data=df_features, x='duration_ms', y='genre', palette='tab20c')

plt.title('Total Duration by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Duration (ms)')

plt.show()
```

```python
genre_duration = df_features.groupby('genre')['popularity'].sum().reset_index()

top_genres = genre_duration.sort_values(by='popularity', ascending=False).head(5)

sns.barplot(data=top_genres, y='popularity', x='genre', palette='RdBu')

plt.title('Top 5 Genres by Popularity')
plt.xlabel('Genres')
plt.ylabel('Popularity')
```
